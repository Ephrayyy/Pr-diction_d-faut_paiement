## Classification binaire avec régression logistique

Je vais utiliser la regression logistique 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

Il existe plusieurs métriques dévaluation du modèle :

#Accuracy 
#Matrice de confusion 
#Rapport de classification 

## Évaluation des performances : Matrice de confusion, AUC, F1

```python
# Accuracy 
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# AUC-ROC

from sklearn.metrics import roc_auc_score, roc_curve
auc_score = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print("AUC-ROC Score :", auc_score)
```

```python
# Rapport de classification 

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Avec ce modèle, on voit que la classification n'est pas précise et prédit de mauvaises données, je vais donc essayer de changer de modèle de classification binaire et passer sur du AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
 
# Utiliser 'estimator' au lieu de 'base_estimator'
base_estimator = DecisionTreeClassifier(max_depth=1)

# Modèle AdaBoost
adaboost_model = AdaBoostClassifier(estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)

# Entraînement
adaboost_model.fit(X_train, y_train)


```

```python
# Prédictions sur le jeu de test
y_pred_adaboost = adaboost_model.predict(X_test)

# Évaluation des performances
print(classification_report(y_test, y_pred_adaboost))
print("Accuracy :", accuracy_score(y_test, y_pred_adaboost))
```

## Interprétation des probabilités pour la prise de décision

1. Compréhension des probabilités
Lorsque vous utilisez predict_proba avec des modèles comme RandomForestClassifier, il retourne des probabilités pour chaque classe.

Exemple :

```python
# Prédictions avec probabilités
y_proba = random_forest_model.predict_proba(X_test)
y_proba
```

Chaque ligne de y_proba contient deux probabilités :
Colonne 0 : La probabilité d’appartenance à la classe 0 (pas de défaut).
Colonne 1 : La probabilité d’appartenance à la classe 1 (défaut de paiement).

2. Seuil de classification par défaut
La plupart des classifieurs utilisent un seuil de 0,5 pour choisir la classe :

Si la probabilité pour la classe 1 est > 0,5, le modèle prédit la classe 1.
Sinon, il prédit la classe 0.

3. Modifier le seuil de classification pour ajuster les performances
Pour équilibrer le précision et le rappel :

Si le rappel pour la classe 1 est trop faible (comme dans votre cas), vous pouvez abaisser le seuil pour la classe 1.

Exemple :

```python
# Utilisation d'un seuil personnalisé
custom_threshold = 0.4
y_pred_custom = (y_proba[:, 1] >= custom_threshold).astype(int)
y_pred_custom
```

4. Stratégies de décision en fonction des probabilités
Stratégie conservatrice (réduire le risque de faux positifs)
Utiliser un seuil plus élevé, par exemple 0.6.
Approprié si les conséquences des faux positifs (prédire un défaut à tort) sont coûteuses.
Stratégie agressive (augmenter le rappel)
Utiliser un seuil plus bas, par exemple 0.3.
Approprié si les faux négatifs (ne pas détecter un défaut réel) sont critiques.

5. Choisir un seuil optimal
Utilisez la courbe ROC pour choisir un seuil qui maximise l’équilibre entre précision et rappel.

Exemple pour tracer la courbe ROC :



```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba[:, 1])

# Visualiser la courbe ROC pour trouver le meilleur seuil
import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

# Conclusion
L’utilisation des probabilités avec des seuils personnalisés est une méthode puissante pour améliorer la prise de décision dans votre modèle. En ajustant le seuil, vous pouvez contrôler le compromis entre précision et rappel, en fonction des besoins métier.
