## Classification binaire avec rÃ©gression logistique

Je vais utiliser la regression logistique 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

Il existe plusieurs mÃ©triques dÃ©valuation du modÃ¨le :

#Accuracy 
#Matrice de confusion

## Ã‰valuation des performances : Matrice de confusion, AUC, F1

```python
# Accuracy 
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# AUC-ROC

from sklearn.metrics import roc_auc_score, roc_curve
auc_score = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print("AUC-ROC Score :", auc_score)
```

```python
# Rapport de classification 

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Avec ce modÃ¨le, on voit que la classification n'est pas prÃ©cise et prÃ©dit de mauvaises donnÃ©es, je vais donc essayer de changer de modÃ¨le de classification binaire et passer sur du AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
 
# Utiliser 'estimator' au lieu de 'base_estimator'
base_estimator = DecisionTreeClassifier(max_depth=1)

# ModÃ¨le AdaBoost
adaboost_model = AdaBoostClassifier(estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)

# EntraÃ®nement
adaboost_model.fit(X_train, y_train)


```

```python
# PrÃ©dictions sur le jeu de test
y_pred_adaboost = adaboost_model.predict(X_test)

# Ã‰valuation des performances
print(classification_report(y_test, y_pred_adaboost))
print("Accuracy :", accuracy_score(y_test, y_pred_adaboost))
```

## InterprÃ©tation des probabilitÃ©s pour la prise de dÃ©cision

1. ComprÃ©hension des probabilitÃ©s
Lorsque vous utilisez predict_proba avec des modÃ¨les comme RandomForestClassifier, il retourne des probabilitÃ©s pour chaque classe.

Exemple :

```python
# PrÃ©dictions avec probabilitÃ©s
y_proba = random_forest_model.predict_proba(X_test)
y_proba
```

Chaque ligne de y_proba contient deux probabilitÃ©s :
Colonne 0 : La probabilitÃ© dâ€™appartenance Ã  la classe 0 (pas de dÃ©faut).
Colonne 1 : La probabilitÃ© dâ€™appartenance Ã  la classe 1 (dÃ©faut de paiement).

2. Seuil de classification par dÃ©faut
La plupart des classifieurs utilisent un seuil de 0,5 pour choisir la classe :

Si la probabilitÃ© pour la classe 1 est > 0,5, le modÃ¨le prÃ©dit la classe 1.
Sinon, il prÃ©dit la classe 0.

3. Modifier le seuil de classification pour ajuster les performances
Pour Ã©quilibrer le prÃ©cision et le rappel :

Si le rappel pour la classe 1 est trop faible (comme dans votre cas), vous pouvez abaisser le seuil pour la classe 1.

Exemple :

```python
# Utilisation d'un seuil personnalisÃ©
custom_threshold = 0.4
y_pred_custom = (y_proba[:, 1] >= custom_threshold).astype(int)
y_pred_custom
```

4. StratÃ©gies de dÃ©cision en fonction des probabilitÃ©s
StratÃ©gie conservatrice (rÃ©duire le risque de faux positifs)
Utiliser un seuil plus Ã©levÃ©, par exemple 0.6.
AppropriÃ© si les consÃ©quences des faux positifs (prÃ©dire un dÃ©faut Ã  tort) sont coÃ»teuses.
StratÃ©gie agressive (augmenter le rappel)
Utiliser un seuil plus bas, par exemple 0.3.
AppropriÃ© si les faux nÃ©gatifs (ne pas dÃ©tecter un dÃ©faut rÃ©el) sont critiques.

5. Choisir un seuil optimal
Utilisez la courbe ROC pour choisir un seuil qui maximise lâ€™Ã©quilibre entre prÃ©cision et rappel.

Exemple pour tracer la courbe ROC :



```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_proba[:, 1])

# Visualiser la courbe ROC pour trouver le meilleur seuil
import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
